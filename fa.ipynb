```python
from pythomata import SimpleNFA, SimpleDFA
from IPython.display import Image # to get FA as image
```

```python
alphabet = {'0', '1'}
states = {'q0', 'q1', 'q2', 'q3'}
initial_state = 'q0'
accepting_states = {'q3'}
transition_function = {
    'q0' : {
        '0' : {'q0', 'q1'},
        '1' : {'q0', 'q2'}
    },
    'q1' : {
        '0' : {'q3'}
    },
    'q2' : {
        '1' : {'q3'}
    }
}    
```

```python
nfa = SimpleNFA(states, alphabet, initial_state, accepting_states, transition_function)

picture = nfa.to_graphviz()
picture.format = 'png'
Image(filename=picture.render("./pictures/new_nfa")) 
```

```python
#dfa = nfa.determinize() # WTF???
dfa = nfa.determinize().trim()

picture = dfa.to_graphviz()
picture.format = 'png'
Image(filename=picture.render("./pictures/new_dfa")) 
```

```python
mdfa = dfa.minimize()

picture = mdfa.to_graphviz()
picture.format = 'png'
Image(filename=picture.render("./pictures/new_mdfa"))
```
